```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

def f(x):
    return np.log(x)

x = np.linspace(0.1, 10, 100)

plt.plot(x,f(x))

p1 = 2, f(2)
p2 = 5, f(5)
points = np.array([p1,p2]).T
plt.scatter(points[0], points[1], marker='o')

dy = p2[1] - p1[1]
dx = p2[0] - p1[0]

p3 = p1[0] + dx, p1[1]

plt.scatter(*p3)


```
