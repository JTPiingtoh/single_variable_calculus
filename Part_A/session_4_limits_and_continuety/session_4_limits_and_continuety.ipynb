```python
def function(x:np.array, a, b):

    y= np.copy(x)

    poly = MultiPoly([1,0,1])

    mask = x > 1
    
    y[mask] = poly.f(x[mask])
    y[~mask] = a*x[~mask] + b


    return y
```

```python
def derivative(x: np.array, a, b):

    y = np.copy(x)
    poly = MultiPoly([1,0,1])
    deriv = MultiPoly(poly.derive())
    mask = x > 1

    y[mask] = deriv.f(x[mask])
    y[~mask] = a*x[~mask]

    return y
```

```python
x = np.linspace(-2,4, 100)

y = function(x, 2, 0)
deriv_y = derivative(x,2,0)

fig, ax = plt.subplots()

ax.plot(x,y)
ax.plot(x,deriv_y)
```

So for $ a = 2, b = 0 $, the function has a derivative
